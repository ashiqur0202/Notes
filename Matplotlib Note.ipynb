# Bar plot

```python
name = ['python','R','C','C++','java']
rank = [85,70,60,86,58]
```

```python
plt.bar(name, rank)
plt.show()
```

```python
plt.bar(name, rank)
plt.xlabel('Programming Language')
plt.ylabel('Rank')
plt.title('Programming Language Popularity')
plt.show()
```

```python
plt.bar(name, rank)
plt.xlabel('Programming Language',fontsize=15)
plt.ylabel('Rank',fontsize=15)
plt.title('Programming Language Popularity',fontsize=15)
plt.show()
```

```python
# with Width paremeter
plt.bar(name, rank, width=0.6)
plt.xlabel('Programming Language',fontsize=15)
plt.ylabel('Rank',fontsize=15)
plt.title('Programming Language Popularity',fontsize=20)
plt.show()
```

```python
name = ['python','R','C','C++','java']
rank = [85,70,60,86,58]
rank2 = [23,46,33,44,56]
```

```python
plt.xlabel('Programming langauge')
plt.ylabel('rank')
plt.title('programming langauge rank')

width = 0.4
p = np.arange(len(name))
p1 = [j+width for j in p]

plt.bar(p,rank,width,color='m',label='rank 2020')
plt.bar(p1,rank2,width, color='g', label='rank 2021')

plt.xticks(p+width/2,name,rotation=90)
plt.legend()
plt.show()
```

```python
plt.ylabel('Programming langauge')
plt.xlabel('rank')
plt.title('programming langauge rank')

width = 0.4
p = np.arange(len(name))
p1 = [j+width for j in p]

plt.barh(p,rank, width, color='m',label='rank 0')
plt.barh(p1,rank2, width, color='g', label='rank 2')

plt.yticks(p+width/2,name,rotation=30)
plt.legend()
plt.show()
```

```python
  
Women = [115, 215, 250, 200]
Men = [114, 230, 510, 370]
  
n=4
r = np.arange(n)
width = 0.3
  
  
plt.bar(r, Women, width, color = 'b', edgecolor = 'black',label='Women')
plt.bar(r + width, Men, width, color = 'g', edgecolor = 'black',label='Men')
  
plt.xlabel("Year")
plt.ylabel("Number of people voted")
plt.title("Number of people voted in each year")
  
    
plt.xticks(r+width/2 ,['2018','2019','2020','2021'])
plt.legend()
  
plt.show()
```

```python
company =['Google','AMAZ','FB','Microsoft']
revenue = [64374,54445,98533,65434]
profit = [56431,34346,76864,23232]

xpos = np.arange(len(company))
```

```python
plt.bar(xpos,revenue,color='r',width=0.4,label='revenue')
plt.bar(xpos+0.4,profit,color='g',width=0.4,label='profit')
plt.ylabel('revenue')

plt.xticks(xpos,company)
plt.legend()
plt.show()
```

```python
w=0.2
plt.barh(xpos,revenue, w, color='r', label='revenue')
plt.barh(xpos+w,profit, w, color='g', label='profit')
plt.ylabel('revenue')

plt.yticks(xpos,company)
plt.legend()
plt.show()
```

# Pie Chart

```python
pie1 = [10,20,30,40]
pie2 = ['C','C++','Java','Python']
```

```python
plt.pie(pie1)
plt.show()
```

```python
# explode,color perameter
c = ['m','g','y','b','r']
ex = [10,10,10,10]
plt.pie(pie1,labels=pie2,explode=ex,colors=c)
plt.title('Pie Chart')
plt.show()
```

```python
# radius perameter
plt.pie(pie1,labels=pie2,autopct= '%0.2f%%',shadow=True,radius=.8)
plt.show()
```

```python
# labeldistance perameter
plt.pie(pie1,labels=pie2,autopct= '%0.2f%%',
        shadow=True,radius=1.2,labeldistance=1.2,
        textprops={'fontsize':15},wedgeprops=
       {'edgecolor':'g','linewidth':5})
plt.show()
```

# Stack plot and Area plot 

```python
x = [1,2,3,4,5]
area1 = [2,3,2,6,5]
area2 = [3,4,5,2,7]
area3 = [4,3,2,6,7]

plt.stackplot(x,area1)

plt.show()
```

```python
# label perameter

l = ['Area1','Area2','Area3']

plt.stackplot(x,area1,area2, area3, labels=l)

plt.legend(loc='upper left')
plt.show()
```

```python
# color perameter

l = ['Area1','Area2','Area3']
c = ['m','r','g']

plt.stackplot(x,area1,area2, area3, labels=l,colors=c)

plt.legend(loc='upper left')
plt.show()
```

```python
# baseline perameter


l = ['Area1','Area2','Area3']
c = ['m','r','g']

plt.stackplot(x,area1,area2, area3, labels=l,colors=c,baseline='sym')

plt.legend(loc='upper left')
plt.show()
```

```python
# baseline perameter


l = ['Area1','Area2','Area3']
c = ['m','r','g']

plt.stackplot(x,area1,area2, area3, labels=l,colors=c,baseline='wiggle')

plt.legend(loc='upper left')
plt.show()
```

```python
# weighted_wiggle

# baseline perameter


l = ['Area1','Area2','Area3']
c = ['m','r','g']

plt.stackplot(x,area1,area2, area3, labels=l,colors=c,baseline='weighted_wiggle')
plt.title('Python Stack Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.legend(loc=3)
plt.show()
```
